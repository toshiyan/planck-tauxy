```python
import numpy as np, curvedsky, prjlib, plottools as pl, quad_func, binning as bn, tools_qrec
from matplotlib.pyplot import *
```

```python
pl.plot_1dstyle(fsize=[20,4],ymin=1e-7,ymax=3e-6,xmin=.5,xmax=2048)
for dtype, wtype, co in [('dr2_smica','Lmask','r'),('dr2_nilc','Lmask','c'),('dr3_nosz','LmaskDR3','b')]:
    p = prjlib.init_analysis(fltr='cinv',dtype=dtype,wtype=wtype)
    qobj = tools_qrec.init_quad(p.ids,p.stag,rlmin=100,rlmax=2048)[3]
    l, al = np.loadtxt(qobj.f['TT'].al,unpack=True,usecols=(0,1))
    l, n0 = np.loadtxt(qobj.f['TT'].n0bs,unpack=True,usecols=(0,1))
    plot(l,al,'--',color=co,label='')
    plot(l,n0,'-',color=co,label=dtype+','+wtype)
legend()
```
